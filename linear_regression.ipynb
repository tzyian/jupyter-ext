```python
# Explain what this cell computes
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression

```

```python
# Generate synthetic data
np.random.seed(0)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)
data = pd.DataFrame(data=np.hstack((X, y)), columns=['Feature', 'Target'])
```

```python
# Fit linear regression model
model = LinearRegression()
model.fit(data[['Feature']], data['Target'])
```

```python
# Make predictions
data['Predicted'] = model.predict(data[['Feature']])
```

```python
# Plot results
plt.scatter(data['Feature'], data['Target'], color='blue', label='Actual Data')
plt.plot(data['Feature'], data['Predicted'], color='red', linewidth=2, label='Regression Line')
plt.xlabel('Feature')
plt.ylabel('Target')
plt.title('Linear Regression Example')
plt.legend()
plt.show()
```
